## Part 2 | Trying OpenCV's built in methods

```python
import matplotlib.pyplot as plt

def display_image(image, title):
    plt.figure(figsize=(10, 5))
    plt.imshow(image)
    plt.title(title)
    plt.axis("off")
    plt.show()
```

```python
import cv2 
image_paths=['assets/stitching/pano_image1.jpg','assets/stitching/pano_image2.jpg'] 

imgs = [] 
for i in range(len(image_paths)): 
	imgs.append(cv2.cvtColor(cv2.imread(image_paths[i]), cv2.COLOR_BGR2RGB))   
	# imgs[i]=cv2.resize(imgs[i],(0,0),fx=0.4,fy=0.4)  #optional if input images aren't too large 

stitchy=cv2.Stitcher.create() 
(dummy,output)=stitchy.stitch(imgs) 

if dummy != cv2.STITCHER_OK: 
# .stitch() function returns a true value if stitching is 
	print("stitching wasn't successful") 
else: 
	print('Your Panorama is ready!!') 

for i in range(len(image_paths)):
    display_image(cv2.cvtColor(cv2.imread(image_paths[i]), cv2.COLOR_BGR2RGB), f"Image {i+1}")
display_image(output, "Final Result")

```
